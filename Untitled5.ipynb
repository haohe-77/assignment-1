```python
from pandas_datareader.data import DataReader

df = DataReader("CPIAUCSL", "fred", "2015-01-01", "2024-12-01")
print(df.head())
print(df.tail())
```

```python
data = DataReader(
    ["CPIAUCSL", "CORESTICKM159SFRBATL", "UNRATE", "FEDFUNDS"],
    "fred",
    "2000-01-01",
    "2024-12-01"
)
```

```python
df = data.copy()

df["inflation"] = df["CPIAUCSL"].pct_change(12) * 100
df["core_inflation"] = df["CORESTICKM159SFRBATL"]
df["unrate"] = df["UNRATE"]
df["fedfunds"] = df["FEDFUNDS"]

df = df.dropna()

print(df.head())
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df["inflation"], label="Inflation")
plt.plot(df.index, df["core_inflation"], label="Core Inflation")
plt.legend()
plt.title("Inflation vs Core Inflation")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

dataset = df[["inflation", "core_inflation", "unrate", "fedfunds"]].values

scaler = MinMaxScaler()
dataset = scaler.fit_transform(dataset)

def create_dataset(dataset, look_back=12):
    X, y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:i+look_back, :])
        y.append(dataset[i+look_back, 0])
    return np.array(X), np.array(y)

look_back = 24

X, y = create_dataset(dataset, look_back)

train_size = int(len(X) * 0.7)

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]

print(X_train.shape, X_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

model = Sequential()
model.add(LSTM(50, input_shape=(12, 4)))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(X_train, y_train, epochs=20, batch_size=16, verbose=2)
```

```python
trainPredict = model.predict(X_train)
testPredict = model.predict(X_test)

y_train_true = y_train.reshape(-1, 1)
y_test_true = y_test.reshape(-1, 1)

dummy_train_pred = np.zeros((len(trainPredict), 4))
dummy_test_pred = np.zeros((len(testPredict), 4))
dummy_train_true = np.zeros((len(y_train_true), 4))
dummy_test_true = np.zeros((len(y_test_true), 4))

dummy_train_pred[:, 0] = trainPredict[:, 0]
dummy_test_pred[:, 0] = testPredict[:, 0]
dummy_train_true[:, 0] = y_train_true[:, 0]
dummy_test_true[:, 0] = y_test_true[:, 0]

trainPredict_inv = scaler.inverse_transform(dummy_train_pred)[:, 0]
testPredict_inv = scaler.inverse_transform(dummy_test_pred)[:, 0]
y_train_inv = scaler.inverse_transform(dummy_train_true)[:, 0]
y_test_inv = scaler.inverse_transform(dummy_test_true)[:, 0]

from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test_inv, testPredict_inv))
mae = mean_absolute_error(y_test_inv, testPredict_inv)

print("Test RMSE:", rmse)
print("Test MAE:", mae)

plt.figure(figsize=(10,5))
plt.plot(y_test_inv, label="True Inflation")
plt.plot(testPredict_inv, label="Predicted Inflation")
plt.legend()
plt.title("LSTM Inflation Prediction")
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(12, 4)))
model.add(Dropout(0.2))
model.add(LSTM(32))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(X_train, y_train, epochs=30, batch_size=16, verbose=2)
```

```python
trainPredict = model.predict(X_train)
testPredict = model.predict(X_test)

y_train_true = y_train.reshape(-1, 1)
y_test_true = y_test.reshape(-1, 1)

dummy_train_pred = np.zeros((len(trainPredict), 4))
dummy_test_pred = np.zeros((len(testPredict), 4))
dummy_train_true = np.zeros((len(y_train_true), 4))
dummy_test_true = np.zeros((len(y_test_true), 4))

dummy_train_pred[:, 0] = trainPredict[:, 0]
dummy_test_pred[:, 0] = testPredict[:, 0]
dummy_train_true[:, 0] = y_train_true[:, 0]
dummy_test_true[:, 0] = y_test_true[:, 0]

trainPredict_inv = scaler.inverse_transform(dummy_train_pred)[:, 0]
testPredict_inv = scaler.inverse_transform(dummy_test_pred)[:, 0]
y_train_inv = scaler.inverse_transform(dummy_train_true)[:, 0]
y_test_inv = scaler.inverse_transform(dummy_test_true)[:, 0]

from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test_inv, testPredict_inv))
mae = mean_absolute_error(y_test_inv, testPredict_inv)

print("Test RMSE:", rmse)
print("Test MAE:", mae)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(y_test_inv, label="True Inflation")
plt.plot(testPredict_inv, label="Predicted Inflation")
plt.legend()
plt.title("Improved LSTM Prediction")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X_flat = X.reshape(X.shape[0], -1)

X_train_flat = X_flat[:train_size]
X_test_flat = X_flat[train_size:]

model_lr = LinearRegression()
model_lr.fit(X_train_flat, y_train)

y_pred_lr = model_lr.predict(X_test_flat)

y_pred_lr = y_pred_lr.reshape(-1,1)
y_test_true = y_test.reshape(-1,1)

dummy_pred = np.zeros((len(y_pred_lr), 4))
dummy_true = np.zeros((len(y_test_true), 4))

dummy_pred[:,0] = y_pred_lr[:,0]
dummy_true[:,0] = y_test_true[:,0]

y_pred_lr_inv = scaler.inverse_transform(dummy_pred)[:,0]
y_test_inv = scaler.inverse_transform(dummy_true)[:,0]

from sklearn.metrics import mean_squared_error
import numpy as np

rmse_lr = np.sqrt(mean_squared_error(y_test_inv, y_pred_lr_inv))

print("Linear Regression RMSE:", rmse_lr)
```
